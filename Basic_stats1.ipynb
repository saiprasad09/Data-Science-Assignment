## Calculate the mean, median, mode, and standard deviation for these columns.

```python
import numpy as np
```

```python
for i in num_col:
  mean = np.mean(df[i])
  median = np.median(df[i])
  mode = df[i].mode()
  std_dev = np.std(df[i])
  print("mean of",i,"is",mean,)
  print("median of",i,"is",median,)
  print("mode of",i,"is",mode,"\n")
  print("standard deviation of",i,"is",std_dev)
```

## 	Create boxplots for numerical variables to identify outliers and the interquartile range.
Discuss any findings, such as extreme values or unusual distributions.


```python
for i in num_col:
  plt.figure(figsize=(3,3))
  plt.boxplot(df[i])
  plt.title(i)
  plt.show()
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3 - Q1
  print ("IQR of ",i ,"is",IQR)
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  # Identify outliers
  outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]
  print("Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
  print("Lower Boundary: {lower_bound}, Upper Boundary: {upper_bound}")
  print("Outliers:\n", outliers)
```

the points outside of the extreme value are outliers in boxplot
