```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Cardiotocographic.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
mean = df['LB'].mean()
df['LB'] = df['LB'].fillna(mean)
```

```python
mean = df['AC'].mean()
df['AC'] = df['AC'].fillna(mean)
```

```python
# filled the mean where there where null values
mean = df['DS'].mean()
df['DS'] = df['DS'].fillna(mean)
mean = df['DP'].mean()
df['DP'] = df['DP'].fillna(mean)
mean = df['MLTV'].mean()
df['MLTV'] = df['MLTV'].fillna(mean)
mean = df['Width'].mean()
df['Width'] = df['Width'].fillna(mean)
mean = df['Tendency'].mean()
df['Tendency'] = df['Tendency'].fillna(mean)
mean = df['NSP'].mean()
df['NSP'] = df['NSP'].fillna(mean)
```

```python
df.info()
```

```python
# as we can see there are no incorrect data types
#now check the duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(keep='first',ignore_index=True,inplace = True)
```

```python
#the duplicates are removed
df.dtypes
```

```python
Q3=df[['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP']].quantile(0.75)
Q3
```

```python
Q1 = df[['LB','AC','FM','UC','DL','DS','DP','ASTV','MSTV','ALTV','MLTV','Width','Tendency','NSP']].quantile(0.25)
```

```python
Q1
```

```python
IQR = (Q3-Q1)
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
print('lower_limit:',lower_limit)
print('upper_limit:',upper_limit)
```

```python
outliers_LB = df[(df['LB']<lower_limit['LB'])| (df['LB']>upper_limit['LB'])]
outliers_LB
outliers_AC = df[(df['AC']<lower_limit['AC'])| (df['AC']>upper_limit['AC'])]
outliers_AC
outliers_FM = df[(df['FM']<lower_limit['FM'])| (df['FM']>upper_limit['FM'])]
outliers_FM
outliers_UC = df[(df['UC']<lower_limit['UC'])| (df['UC']>upper_limit['UC'])]
outliers_UC
outliers_DL = df[(df['DL']<lower_limit['DL'])| (df['DL']>upper_limit['DL'])]
outliers_DL
outliers_DS = df[(df['DS']<lower_limit['DS'])| (df['DS']>upper_limit['DS'])]
outliers_DS
outliers_DP = df[(df['DP']<lower_limit['DP'])| (df['DP']>upper_limit['DP'])]
outliers_DP
outliers_ASTV = df[(df['ASTV']<lower_limit['ASTV'])| (df['ASTV']>upper_limit['ASTV'])]
outliers_ASTV
outliers_MSTV = df[(df['MSTV']<lower_limit['MSTV'])| (df['MSTV']>upper_limit['MSTV'])]
outliers_MSTV
outliers_ALTV = df[(df['ALTV']<lower_limit['ALTV'])| (df['ALTV']>upper_limit['ALTV'])]
outliers_ALTV
outliers_MLTV = df[(df['MLTV']<lower_limit['MLTV'])| (df['MLTV']>upper_limit['MLTV'])]
outliers_MLTV
outliers_Width = df[(df['Width']<lower_limit['Width'])| (df['Width']>upper_limit['Width'])]
outliers_Width
outliers_Tendency = df[(df['Tendency']<lower_limit['Tendency'])| (df['Tendency']>upper_limit['Tendency'])]
outliers_Tendency
outliers_NSP = df[(df['NSP']<lower_limit['NSP'])| (df['NSP']>upper_limit['NSP'])]
outliers_NSP
```

```python
df = df[(df[['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']] >= lower_limit[['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']]) & (df[['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']] <= upper_limit[['LB', 'AC', 'FM', 'UC', 'DL', 'DS', 'DP', 'ASTV', 'MSTV', 'ALTV', 'MLTV', 'Width', 'Tendency', 'NSP']])]
#inplace=True is not valid for this operation, the assignment must be done explicitly.
```
